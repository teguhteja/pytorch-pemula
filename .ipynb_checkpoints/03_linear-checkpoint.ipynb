```python
import torch
```

```python
w = torch.tensor(3.0, requires_grad=True) 
b = torch.tensor(1.0, requires_grad=True) 
```

```python
def forward(x):
    y = w*x+b
    return y
```

```python
x = torch.tensor([[4],[5.5]])
```

```python
forward(x)
```

```python
from torch.nn import Linear
```

```python
model = Linear(in_features=1, out_features=1)
print(model.weight)
print(model.bias)
```

```python
model(x)
```
