```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
y= x**2 + x -6
```

```python
y.backward()
x.grad
```

```python
x = torch.tensor(2.0, requires_grad=True)
z = torch.tensor(3.0, requires_grad=True)
y = x**2 + z
```

```python
y.backward()
x.grad
```
